# README.md: 競合製品データ分析による市場競争力強化

## プロジェクト概要

本プロジェクトは、`competitors_raw_data.csv`データセットに含まれる競合製品の情報を詳細に分析し、自社製品の市場競争力強化を目的とした戦略的インサイトを導出することを目的としています。価格、評価、機能、カテゴリといった多角的な視点から競合製品を評価し、KGI（Key Goal Indicator）である「自社製品の市場競争力強化」に貢献する具体的なアクションプランを提案します。

## データ概要

### データセット: `competitors_raw_data.csv`

このデータセットは、競合製品に関する以下の8つの列を持つ247行のデータで構成されています。

*   **`Title`**: 製品のタイトル
*   **`Price`**: 製品の価格
*   **`Category_path`**: 製品が属するカテゴリパス（階層構造）
*   **`Rating`**: 顧客による製品の平均評価スコア
*   **`Qty_califications`**: 製品の評価数
*   **`Features_JSON_format`**: 製品の機能リスト（JSON形式）
*   **`Image_Url`**: 製品画像のURL
*   **`Item_Url`**: 製品詳細ページのURL

### データ理解の要点:

*   **欠損値**: `Rating`, `Qty_califications`, `Features_JSON_format`に欠損値が存在し、これらはデータクリーニングフェーズで適切に処理されました。
    *   `Rating`と`Qty_califications`は中央値で補完。
    *   `Features_JSON_format`は空のJSONオブジェクト文字列で補完。
*   **外れ値**: `Price`, `Rating`, `Qty_califications`列においてIQR法を用いた外れ値が検出されました。これらはビジネス上重要な意味を持つ可能性があり、分析では維持されました。
*   **異常値**: 業務ロジックに基づく異常値（例: 価格が0以下）は検出されませんでした。
*   **データ品質スコア**: データ品質スコアは96.96点と非常に高く、分析結果の信頼性が高いことを示唆しています。

## 分析手法

本分析は、以下の主要なステップで構成されています。

1.  **ビジネス理解**: KGI（Key Goal Indicator）、KPI（Key Performance Indicator）、およびKPIツリーを定義し、分析対象を明確化しました。
    *   **KGI**: 自社製品の市場競争力強化
    *   **KPI**: 競合製品の平均価格、平均評価スコア、平均評価数、主要カテゴリにおける競合製品数、特定機能を持つ競合製品の割合
    *   **KPIツリー**: KGIとKPIの因果関係をmermaid形式で視覚化し、相互関係を整理しました。

2.  **データクレンジングと前処理**: 欠損値の処理、データ型変換（`Features_JSON_format`をJSONオブジェクトへ）、外れ値・異常値の検出を行いました。

3.  **探索的データ分析 (EDA)**: `Price`, `Rating`, `Qty_califications`, `Category_path`, `Features_JSON_format`の各列について、記述統計量の算出、ヒストグラムや箱ひげ図による視覚化、上位カテゴリ・機能の分析を実施しました。

4.  **仮説検証**: EDAで生成された3つの仮説を統計的に検証しました。
    *   **仮説 1**: 価格と評価スコアの関係性
    *   **仮説 2**: 評価数と人気機能の関係性
    *   **仮説 3**: カテゴリごとの市場特性

5.  **戦略策定**: 検証された仮説と効果量に基づき、具体的なアクションプラン（価格設定、機能開発、品質向上、市場選定）を策定しました。

6.  **Impact x Effort分析**: 策定した戦略に対してImpact（影響度）とEffort（労力）の評価基準を適用し、実行優先順位を決定しました。

## 主要インサイト

### EDAからの洞察:

*   **価格分布**: 低価格帯に多くの製品が集中する一方で、高価格帯にもニッチな製品が存在。市場は価格多様性が高い。
*   **評価スコア分布**: 全体的に高評価の製品が多く、市場の顧客満足度は高い水準にある。低評価の製品も少数存在するが、多くの製品が高品質を提供している。
*   **評価数分布**: 多くの製品は評価数が少ないが、一部の製品は非常に多くの評価を集め、高い人気と認知度を持つ。
*   **カテゴリ多様性**: データセットは非常に細分化されたカテゴリパスを持ち、特定のカテゴリに製品が集中する傾向は見られない。市場の細分化が進んでいることを示唆。
*   **機能トレンド**: 「Material」「Descripcion_0」「Anchura exterior」といった基本的な構成要素や情報に関する機能が最も頻繁に出現する。
*   **カテゴリ別比較**: 上位カテゴリ間で価格、評価スコア、評価数に傾向の違いが見られ、カテゴリごとに異なる市場特性や競争環境が存在。

### 仮説検証結果からのインサイト:

*   **仮説1: 価格と評価**: 「高価格帯の競合製品が低価格帯の製品より平均評価スコアが統計的に有意に高い」という仮説は支持されませんでした（P値: 0.171, Cohen's d: -0.176）。これは、単に価格を高くするだけでは評価に直結しないことを示唆しています。価格以外の要因（品質、機能、ブランドなど）が評価に大きく影響している可能性があります。
*   **仮説2: 評価数と人気機能**: 「人気製品群は特定の機能の搭載割合が有意に高い」という仮説は、**Material**, **Altura total**, **Anchura interior**, **Número de baldas**, **Grueso de laterales y baldas**, **Fondo estantería** の機能において統計的に有意な差が見られましたが、Cramer's Vによる効果量は小さい傾向にありました。これは、これらの機能が人気獲得に寄与する可能性はあるものの、決定的な要因ではないことを示唆します。製品の人気は複合的な要因で決定されると考えられます。
*   **仮説3: カテゴリ別市場特性**: 選定された2つのカテゴリ（'Cama 90 con cajón y colchón'と'Estructura de cama 105'）間では、**Price** (P値: 0.005, Cohen's d: 17.102) と **Rating** (P値: 0.042, Cohen's d: 15.000) において統計的に有意かつ非常に大きな差が確認されました。しかし、**Qty_califications** (P値: 0.223, Cohen's d: 1.800) では有意な差は見られませんでした。これは、カテゴリ間で価格設定や顧客評価の期待値に明確な違いがある一方で、評価数についてはこれらの特定のカテゴリ間では顕著な差がないことを示唆します。

## 施策

検証結果と効果量の分析に基づき、以下の4つの具体的な行動戦略を策定しました。

1.  **価格設定と価値訴求戦略**: 顧客が納得する「価値」に基づいた価格設定を行い、ターゲット市場セグメントに合わせて最適化。高価格帯製品では品質・独自機能・顧客体験を明確に訴求する。
2.  **製品機能の優先順位付けと開発戦略**: 人気製品に共通する基本機能を自社製品に標準装備し、競合より優れた形で提供。同時に、市場の差別化要因となるニッチな機能や革新的な機能の探索・開発を継続する。
3.  **品質向上と顧客体験戦略**: 全価格帯の製品で一定水準以上の品質を確保し、顧客レビュー分析を通じて品質改善に活かす。特に高価格帯製品では、徹底した品質管理と優れた顧客サポートで、高価格を正当化する顧客体験を創出する。
4.  **ターゲット市場とカテゴリ選定戦略**: 競合の「Category_path」データを深く分析し、カテゴリごとの市場特性を把握することで、自社が強みを活かせるニッチ市場や成長市場を特定し、ターゲットカテゴリに特化した戦略を策定する。

## ROI評価と限界

### ROI分析のためのデータ利用可能性:

`competitors_raw_data.csv`データセットには、製品の原価、販売数量、総収益、マーケティング費用といったROI（Return on Investment）分析に不可欠なデータが含まれていません。したがって、このデータセット単独で包括的なROI分析を行うことはできません。

### 代替パフォーマンス指標:

既存のデータから間接的に経済的価値や市場の成功度合いを推測するため、以下の代替指標を提案します。

*   **相対価格競争力**: 製品価格を同一カテゴリ内の競合製品の平均価格と比較。
*   **評価効率**: `Rating × log(Qty_califications + 1)`で評価の質と量を複合的に評価。
*   **人気度指数**: `Qty_califications`を基に市場での相対的な人気度を測る。
*   **プレミアム度合い**: 高価格帯で高評価を獲得している製品の度合いを示す。

### 分析の限界:

*   **データ範囲の制約**: 分析は提供された競合製品データに限定されており、自社製品のデータや市場シェア、財務情報など広範なデータは含まれていません。
*   **サンプルサイズの制約**: カテゴリ別の比較分析では、各カテゴリの製品数が少ないため、統計的検定の信頼性や一般化に限界があります。特に'Qty_califications'の比較では、p値の信頼性が低い可能性が指摘されました。
*   **機能情報の粒度**: `Features_JSON_format`はJSON形式ですが、その内容の均一性や構造にばらつきがあり、詳細な機能比較には追加の前処理が必要となる可能性があります。

## Impact x Effort 分析

### Impact x Effort マトリックス: 行動戦略の優先順位付け

### 戦略の実行優先順位

Impact x Effort分析に基づき、以下の原則で行動戦略の実行優先順位を決定しました。
「Impactが高いほど優先度が高く、同じImpactレベルであればEffortが低いほど優先度が高い」

1.  **品質向上と顧客体験戦略**
    *   **Impact**: High
    *   **Effort**: Medium
    *   **優先順位の理由**: 顧客ロイヤリティとブランドイメージに直接影響し、長期的な市場競争力強化に不可欠であるため、高いインパクトを持ちます。品質管理プロセスの強化や顧客サポート体制の改善には労力がかかるものの、大規模な製品開発に比べれば中程度であるため、早期に着手し、その効果を最大化すべきです。

2.  **製品機能の優先順位付けと開発戦略**
    *   **Impact**: High
    *   **Effort**: High
    *   **優先順位の理由**: 基本機能の改善は顧客満足度向上に直結し、差別化機能は新たな顧客層獲得に繋がり、KGIへの影響は非常に高いです。しかし、機能開発には研究開発、設計、製造工程の変更など多大な時間、予算、技術リソースが必要となるため、最も高い労力を伴います。中長期的な視点での計画とリソース配分が必要であり、初期段階での市場と顧客のニーズのさらなる詳細な調査とプロトタイプ開発から始めることが推奨されます。

3.  **ターゲット市場とカテゴリ選定戦略**
    *   **Impact**: Medium
    *   **Effort**: Low
    *   **優先順位の理由**: 適切なターゲット市場とカテゴリの選定は、リソースの効率的な配分と効果的な戦略策定を可能にし、KGIに中程度の貢献をします。主にデータ分析と戦略立案が中心であり、即座に大きな投資を伴わないため労力は低いです。市場理解を深め、他の戦略の精度を高めるために非常に重要であるため、即座に実行を開始すべきです。

4.  **価格設定と価値訴求戦略**
    *   **Impact**: Medium
    *   **Effort**: Medium
    *   **優先順位の理由**: 価格設定の調整は比較的迅速に可能ですが、価値訴求のためにはマーケティング戦略の見直しやブランドメッセージの強化が必要です。市場セグメント別の最適化にはある程度の市場分析と実行期間を要するため、中程度のインパクトと労力と評価されます。品質向上や市場理解が進んだ段階で、より洗練された形で実施することでその効果を高めることができます。

### 全体的なビジネスへの示唆と推奨されるアプローチ

今回の優先順位付けは、限られたリソースの中で最大の効果を得るための指針となります。特に、「**品質向上と顧客体験戦略**」は高いインパクトを持ちながら労力が中程度であるため、早期に着手し、その効果を最大化すべきです。これと並行して、労力が低い「**ターゲット市場とカテゴリ選定戦略**」は、市場理解を深め、他の戦略の精度を高めるために非常に重要であり、即座に実行を開始すべきです。「**製品機能の優先順位付けと開発戦略**」は高いインパクトが期待されるものの、最も高い労力を伴うため、中長期的な視点での計画とリソース配分が必要です。初期段階での市場と顧客のニーズのさらなる詳細な調査とプロトタイプ開発から始めることが推奨されます。「**価格設定と価値訴求戦略**」は中程度のインパクトと労力であり、品質向上や市場理解が進んだ段階で、より洗練された形で実施することでその効果を高めることができます。

これらの戦略を段階的に、かつ相互に関連付けながら実行することで、持続的な市場競争力の強化が期待できます。


````python
import zipfile
import os
import pandas as pd

# Unzip the archive
zip_path = '/content/archive.zip'
unzip_dir = '/content/unzipped_data'

# Create a directory to extract files into, if it doesn't exist
os.makedirs(unzip_dir, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_dir)

print(f"Archive unzipped to: {unzip_dir}")

# List all CSV files in the unzipped directory
csv_files = [f for f in os.listdir(unzip_dir) if f.endswith('.csv')]
print("Found CSV files:", csv_files)
````

````python
def perform_data_understanding(file_path):
    """Performs data understanding for a given CSV file."""
    print(f"\n--- Data Understanding for {os.path.basename(file_path)} ---")

    # Try different encodings to read the CSV file
    encodings_to_try = ['utf-8', 'shift_jis', 'latin1']
    df = None
    for encoding in encodings_to_try:
        try:
            # Add on_bad_lines='skip' to handle malformed lines
            df = pd.read_csv(file_path, encoding=encoding, on_bad_lines='skip')
            print(f"ファイルは'{encoding}'エンコーディングで正常に読み込まれました。")
            break
        except UnicodeDecodeError:
            print(f"'{encoding}'エンコーディングでの読み込みに失敗しました。次のエンコーディングを試します。")
            continue
        except Exception as e:
            print(f"ファイルの読み込み中に予期せぬエラーが発生しました: {e}")
            return

    if df is None:
        print(f"指定されたすべてのエンコーディングでファイルを読み込むことができませんでした: {file_path}")
        return

    # 1. 行数・列数
    rows, cols = df.shape
    print(f"行数: {rows}")
    print(f"列数: {cols}")

    # 2. データ型一覧
    print("\nデータ型一覧:")
    display(df.dtypes.to_frame(name='データ型'))

    # 3. 欠損率（列別）
    missing_rates = df.isnull().sum() / len(df) * 100
    print("\n欠損率 (%):")
    display(missing_rates[missing_rates > 0].to_frame(name='欠損率'))
    if missing_rates.empty:
        print("欠損値はありません。")

    # 4. 重複数
    duplicate_rows = df.duplicated().sum()
    print(f"\n重複行数: {duplicate_rows}")

    # 5. 主キー候補評価（理由付き）
    print("\n主キー候補評価:")
    pk_candidates = []
    for col in df.columns:
        if df[col].nunique() == rows:
            pk_candidates.append(col)

    if pk_candidates:
        print("以下の列が単独で主キー候補となります（すべての値が一意であるため）:")
        for pk in pk_candidates:
            print(f"- {pk}")
    else:
        print("単独で主キーとなりうる列はありません。複合主キーの可能性があります。")
        # Check for combinations of up to 3 columns as potential primary keys
        from itertools import combinations
        found_composite_pk = False
        for r in range(2, min(4, cols + 1)): # Check combinations of 2 to 3 columns
            for combo in combinations(df.columns, r):
                if df.duplicated(subset=list(combo)).sum() == 0:
                    print(f"- 複合主キー候補: {list(combo)} (組み合わせた値が全て一意であるため)")
                    found_composite_pk = True
                    break # Found one, move to next file or stop for this file
            if found_composite_pk: break
        if not found_composite_pk: print("複合主キー候補も見つかりませんでした。")

    # 6. 分析可能な粒度（1行定義）
    print("\n分析可能な粒度（1行定義）:")
    if pk_candidates:
        print(f"このデータセットの1行は、'{pk_candidates[0]}' の単位での情報を示していると推測されます。具体的には、各行は一意な{pk_candidates[0]}に関連する観測データです。")
    else:
        print("単一の主キー候補が特定できないため、1行の粒度はデータの内容に基づいて慎重に評価する必要があります。複合主キー候補が見つかった場合、その組み合わせが粒度を定義します。")
        print("例: '日時' + '商品ID' で1つのトランザクションを表す、など。")

# Iterate through each CSV file and perform data understanding
for csv_file in csv_files:
    file_path = os.path.join(unzip_dir, csv_file)
    perform_data_understanding(file_path)

print("\nすべてのCSVファイルのデータ理解が完了しました。")
````

## 外れ値検証

### Subtask:
`Price`, `Rating`, `Qty_califications`列に対し、IQR法を用いて外れ値を検出します。外れ値の数と分布を報告し、これらの外れ値がデータ全体に与える潜在的なビジネス影響を考察します。ただし、データ修正は行わず、検出のみに留めます。


**Reasoning**:
Calculate Q1, Q3, IQR, and the outlier bounds for each of the specified columns (`Price`, `Rating`, `Qty_califications`). This is the initial step to identify outliers using the IQR method.



````python
import numpy as np

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f"\n--- {column} 列の外れ値検出 (IQR法) ---")
    print(f"Q1 (第1四分位数): {Q1:.2f}")
    print(f"Q3 (第3四分位数): {Q3:.2f}")
    print(f"IQR (四分位範囲): {IQR:.2f}")
    print(f"下限 (Q1 - 1.5 * IQR): {lower_bound:.2f}")
    print(f"上限 (Q3 + 1.5 * IQR): {upper_bound:.2f}")
    print(f"{column} 列の外れ値数: {len(outliers)}")
    print(f"外れ値の割合: {len(outliers)/len(df)*100:.2f}%")

    if not outliers.empty:
        print(f"{column} 列の外れ値の統計:")
        display(outliers[column].describe())
    else:
        print("外れ値はありません。")

    return outliers

# Detect outliers for 'Price'
price_outliers = detect_outliers_iqr(df_competitors, 'Price')

# Detect outliers for 'Rating'
rating_outliers = detect_outliers_iqr(df_competitors, 'Rating')

# Detect outliers for 'Qty_califications'
qty_califications_outliers = detect_outliers_iqr(df_competitors, 'Qty_califications')
````

## データ型変換

### Subtask:
データフレームの各列について、現在のデータ型を確認し、必要に応じて適切なデータ型に変換します。例えば、`Features_JSON_format`列は文字列として読み込まれていますが、JSON形式としてより扱いやすい形への変換を検討します。変更内容、影響件数、ビジネス影響を報告します。


**Reasoning**:
The next logical step is to convert the `Features_JSON_format` column, which is currently an object (string), into actual JSON objects (Python dictionaries) to make it more amenable for analysis. This directly addresses the subtask's instruction to convert this column to a more usable form.



````python
import json

# Initialize a counter for affected rows
affected_rows_json_conversion = 0

def parse_json_safe(json_str):
    global affected_rows_json_conversion
    try:
        parsed_json = json.loads(json_str)
        # If it was a string and successfully parsed, it's an affected row.
        # This check ensures that already valid JSON-like structures (empty dicts from fillna) aren't counted unnecessarily
        # unless they were originally malformed and fixed.
        # For simplicity, we'll count any successful conversion from a non-dict type.
        if not isinstance(parsed_json, dict):
            affected_rows_json_conversion += 1
        return parsed_json
    except json.JSONDecodeError:
        affected_rows_json_conversion += 1
        return {} # Return an empty dictionary for malformed JSON entries

print("データ型変換前のFeatures_JSON_format列のデータ型:", df_competitors['Features_JSON_format'].dtype)

# Apply the safe JSON parsing function
df_competitors['Features_JSON_format'] = df_competitors['Features_JSON_format'].apply(parse_json_safe)

print("データ型変換後のFeatures_JSON_format列のデータ型:", df_competitors['Features_JSON_format'].dtype)
print(f"変更内容: Features_JSON_format列の文字列データをJSONオブジェクト(Python辞書)に変換しました。")
print(f"影響件数: {affected_rows_json_conversion}件 (元の欠損値補完を除く、解析エラーまたは形式変更による影響)")
print(f"ビジネス影響: 製品の特徴データをプログラムで容易にアクセス、分析できるようになります。これにより、特定の機能を持つ製品の割合を正確に計算したり、機能の傾向を把握したりすることが可能になり、自社製品の機能開発戦略や市場ポジショニングの理解を深めるのに役立ちます。")

# Display the first few entries to confirm the conversion
print("\nFeatures_JSON_format列の変換後の先頭5エントリ:")
display(df_competitors['Features_JSON_format'].head())
````

## データ品質スコアの算出と報告

### Subtask:
データクリーニングプロセスを通じて検出された欠損値、外れ値、異常値の合計に基づき、データセットのデータ品質スコア（0-100）を算出します。スコアの算出根拠と、現在のデータ品質がビジネス意思決定に与える影響について報告します。


**Reasoning**:
First, I will re-confirm the counts of missing values, outliers, and business logic anomalies from the previous subtasks. Then, I will define a data quality scoring method and calculate the score, which addresses the first three instructions of the subtask.



````python
print("--- データ品質スコア算出のための集計 ---")

# 1. 欠損値の総数（処理前のもの）を再確認
initial_missing_rating = 7 # df_competitors['Rating'].isnull().sum() before imputation
initial_missing_qty_califications = 7 # df_competitors['Qty_califications'].isnull().sum() before imputation
initial_missing_features = 9 # df_competitors['Features_JSON_format'].isnull().sum() before imputation

total_initial_missing_values = initial_missing_rating + initial_missing_qty_califications + initial_missing_features
print(f"総欠損値数（処理前）: {total_initial_missing_values}件")

# 2. IQR法による外れ値の総数
# これらの値は'outliers'データフレームの行数として以前に計算済み
total_outliers = len(price_outliers) + len(rating_outliers) + len(qty_califications_outliers)
print(f"IQR法による総外れ値数: {total_outliers}件")

# 3. 業務ロジックに基づく異常値の総数
# これらの値は'abnormal_count'変数として以前に計算済み
total_abnormal_values = abnormal_price_count + abnormal_rating_count + abnormal_qty_califications_count
print(f"業務ロジックに基づく総異常値数: {total_abnormal_values}件")

# データセットの総セル数を計算
total_rows = len(df_competitors)
total_cols = df_competitors.shape[1]
total_cells = total_rows * total_cols
print(f"データセットの総行数: {total_rows}")
print(f"データセットの総列数: {total_cols}")
print(f"データセットの総セル数: {total_cells}")

# データ品質スコアの算出
# 問題件数の合計: 欠損値は修正済みだが、スコア算出の際には"問題"としてカウントする。
# 外れ値と異常値は修正されていないため、そのままカウントする。
# ただし、同じデータポイントが複数カテゴリで問題とカウントされる可能性を考慮し、
# ここでは単純に各カテゴリの件数を合計する。
# もし、より厳密に重複を排除するなら、問題のある行のユニークな数を数えることも考えられる。
# 今回は問題の種類ごとの合計とする。

total_problematic_instances = total_initial_missing_values + total_outliers + total_abnormal_values

# スコア算出ロジックの定義: 100 - (総問題インスタンス数 / 総セル数) * 100
# スコアの最大値が100、問題がない場合に100点となるように調整。
# 問題の"重み"をどう考えるかにもよるが、今回は単純な割合ベースとする。
# なお、"問題インスタンス数"は、各問題カテゴリで検出された件数の合計とする。

# スコアの基準を明確にするために、問題の種類ごとに最大の悪影響を考慮する。
# ここでは、データセット全体に対する問題の"発生数"の割合をペナルティとする。

# より実用的なスコアとして、影響を受けた"行"の割合を考慮する。
# ここでは、簡潔に合計の問題インスタンス数を使用。

data_quality_score = 100 - (total_problematic_instances / total_cells) * 100

print(f"\nデータ品質スコア: {data_quality_score:.2f}点")
````

````python
median_price = df_competitors['Price'].median()
print(f"競合製品の価格の中央値: {median_price:.2f}")

df_competitors['price_range'] = df_competitors['Price'].apply(lambda x: 'High' if x > median_price else 'Low')

print("\n'price_range'カラムを追加後のDataFrameの先頭5行:")
display(df_competitors.head())
````

**Reasoning**:
Perform an independent samples t-test and calculate Cohen's d effect size to compare the mean 'Rating' between 'High' and 'Low' price ranges, as required by the subtask.



````python
from scipy import stats
import numpy as np

# Separate ratings for high and low price ranges
ratings_high_price = df_competitors[df_competitors['price_range'] == 'High']['Rating']
ratings_low_price = df_competitors[df_competitors['price_range'] == 'Low']['Rating']

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(ratings_high_price, ratings_low_price, equal_var=False) # Welch's t-test assuming unequal variances

# Calculate Cohen's d for effect size
def cohens_d(x, y):
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    pooled_std = np.sqrt(((nx-1)*np.std(x, ddof=1) ** 2 + (ny-1)*np.std(y, ddof=1) ** 2) / dof)
    return (np.mean(x) - np.mean(y)) / pooled_std

effect_size_d = cohens_d(ratings_high_price, ratings_low_price)

print("\n--- Hypothesis 1 検証結果 (価格帯別平均評価スコアの比較) ---")
print(f"高価格帯製品の平均評価スコア: {ratings_high_price.mean():.2f}")
print(f"低価格帯製品の平均評価スコア: {ratings_low_price.mean():.2f}")
print(f"T-統計量: {t_statistic:.3f}")
print(f"P-値: {p_value:.3f}")
print(f"Cohen's d (効果量): {effect_size_d:.3f}")

# Interpret p-value and effect size
alpha = 0.05
if p_value < alpha:
    print("統計的有意性: P-値は有意水準 (0.05) より小さいため、統計的に有意な差があります。")
    print("仮説1を支持します: 高価格帯の競合製品は、低価格帯の製品と比較して、平均評価スコアが統計的に有意に高いです。")
else:
    print("統計的有意性: P-値は有意水準 (0.05) より大きいため、統計的に有意な差はありません。")
    print("仮説1を支持しません: 高価格帯の競合製品と低価格帯の製品の間で、平均評価スコアに統計的に有意な差は見られませんでした。")

print("\n効果量の解釈 (Cohen's d):")
if abs(effect_size_d) < 0.2:
    print("効果量d < 0.2: 非常に小さい効果 (実質的な差はほとんどない)")
elif abs(effect_size_d) < 0.5:
    print("効果量0.2 <= d < 0.5: 小さい効果 (わずかな実質的な差がある)")
elif abs(effect_size_d) < 0.8:
    print("効果量0.5 <= d < 0.8: 中程度の効果 (中程度の大きさの実質的な差がある)")
else:
    print("効果量d >= 0.8: 大きい効果 (実質的に大きな差がある)")
````

**Reasoning**:
Perform independent samples t-tests and calculate Cohen's d effect size for 'Price', 'Rating', and 'Qty_califications' between the two selected categories.



````python
from scipy import stats
import numpy as np

def cohens_d(x, y):
    nx = len(x)
    ny = len(y)
    dof = nx + ny - 2
    if dof == 0:
        return np.nan # Avoid division by zero if sample size is too small
    pooled_std = np.sqrt(((nx - 1) * np.std(x, ddof=1) ** 2 + (ny - 1) * np.std(y, ddof=1) ** 2) / dof)
    if pooled_std == 0:
        return np.nan # Avoid division by zero if pooled std is zero
    return (np.mean(x) - np.mean(y)) / pooled_std

def conduct_ttest_and_report(df1, df2, column, category1_name, category2_name):
    data1 = df1[column]
    data2 = df2[column]

    print(f"\n--- {column} のカテゴリ別比較 --- ")
    print(f"平均 {column} ({category1_name}): {data1.mean():.2f}")
    print(f"平均 {column} ({category2_name}): {data2.mean():.2f}")

    # Check if there's enough variance for t-test
    if data1.std() == 0 and data2.std() == 0:
        print(f"両カテゴリの{column}が一定値であり、T-検定を実行できません。")
        if data1.mean() == data2.mean():
            print(f"両カテゴリの平均{column}は同じです。")
        else:
            print(f"両カテゴリの平均{column}は異なりますが、統計的有意性の評価はできません。")
        return

    t_statistic, p_value = stats.ttest_ind(data1, data2, equal_var=False) # Welch's t-test
    effect_size_d = cohens_d(data1, data2)

    print(f"T-統計量: {t_statistic:.3f}")
    print(f"P-値: {p_value:.3f}")
    print(f"Cohen's d (効果量): {effect_size_d:.3f}")

    alpha = 0.05
    if p_value < alpha:
        print(f"統計的有意性: P-値は有意水準 ({alpha}) より小さいため、統計的に有意な差があります。")
        print(f"結論: {category1_name} と {category2_name} の間で、{column} に統計的に有意な差が見られます。")
    else:
        print(f"統計的有意性: P-値は有意水準 ({alpha}) より大きいため、統計的に有意な差はありません。")
        print(f"結論: {category1_name} と {category2_name} の間で、{column} に統計的に有意な差は見られませんでした。")

    print("効果量の解釈 (Cohen's d):")
    if abs(effect_size_d) < 0.2:
        print("非常に小さい効果 (実質的な差はほとんどない)")
    elif abs(effect_size_d) < 0.5:
        print("小さい効果 (わずかな実質的な差がある)")
    elif abs(effect_size_d) < 0.8:
        print("中程度の効果 (中程度の大きさの実質的な差がある)")
    else:
        print("大きい効果 (実質的に大きな差がある)")

category1_name = 'Cama 90 con cajón y colchón'
category2_name = 'Estructura de cama 105'

# Compare Price
conduct_ttest_and_report(df_cat1, df_cat2, 'Price', category1_name, category2_name)

# Compare Rating
conduct_ttest_and_report(df_cat1, df_cat2, 'Rating', category1_name, category2_name)

# Compare Qty_califications
conduct_ttest_and_report(df_cat1, df_cat2, 'Qty_califications', category1_name, category2_name)
````
